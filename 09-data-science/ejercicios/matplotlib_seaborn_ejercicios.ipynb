```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

## Ejericicio 1: Visualizaciones básicas

- Visualiza cuántas personas vivieron vs cuántas murieron en el accidente

```python
ax = sns.countplot(data=titanic, x='survived')
ax.set(title='Supervivientes totales', xlabel='Supervivientes', ylabel='Total')
plt.show()
```

- Comparativa de pasajeros que vivieron o murieron dependiendo de su clase (`pclass`)

```python
sns.countplot(data=titanic, x='pclass', hue='survived')
```

- Comparativa de pasajeros que murieron dependiendo de su género (`sex`)

```python
sns.countplot(data=titanic, x='sex', hue='survived')
```

- Comparativa de pasajeros que murieron o sobrevivieron según la puerta de embarque (`embarked`)

```python
sns.countplot(data=titanic, x='embarked', hue='survived')
```

- Comparativa de pasajeros que murieron dependiendo del número de personas que les acompañaban (`sibsp`)

```python
sns.countplot(data=titanic, x='sibsp', hue='survived')
```

- Comparativa de pasajeros que murieron o sobrevivieron según el rango de edad (`age`)
> Pista 1: prueba a hacer intervalos de edad con `pd.cut`

```python
interval = [0, 18, 35, 60, 120]
categories = ['Children', 'Teens', 'Adult', 'Old']
titanic['age_cats'] = pd.cut(titanic.age, interval, labels=categories)
sns.countplot(data=titanic, x='age_cats', hue='survived')
```

- Comparativa de pasajeros que murieron según el precio del pasaje (`fare`)
> Prueba con un `FacetGrid`

```python
fare = sns.FacetGrid(data=titanic, hue='survived', aspect=3)
fare.map(sns.kdeplot, 'fare', fill=True)
fare.add_legend()
```
